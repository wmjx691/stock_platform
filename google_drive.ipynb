```python
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive

def authenticate_google_drive():
    gauth = GoogleAuth()
    gauth.LocalWebserverAuth()  # 開啟本地 web 伺服器以完成授權
    return GoogleDrive(gauth)

def upload_to_google_drive(file_path, folder_id=None):
    drive = authenticate_google_drive()

    try:
        # 上傳到指定的資料夾，如果未指定則上傳到根目錄
        file_drive = drive.CreateFile({'title': file_path.split("/")[-1], 'parents': [{'id': folder_id}]})
        file_drive.Upload()

        print(f"檔案 {file_path} 成功上傳到 Google 雲端硬碟！")
    except Exception as e:
        print(f"上傳失敗：{e}")
```

```python
def save_to_google_drive(dataframe, file_name, folder_id):
    try:
        # 將資料儲存成 Excel 檔案
        dataframe.to_excel(file_name, index=False)

        # 連接到 Google Drive
        gauth = GoogleAuth()
        gauth.LocalWebserverAuth()
        drive = GoogleDrive(gauth)

        # 將檔案上傳到指定的 Google Drive 資料夾
        file_drive = drive.CreateFile({'title': file_name, 'parents': [{'id': folder_id}]})
        file_drive.Upload()

        print(f"File '{file_name}' uploaded to Google Drive successfully.")
    except Exception as e:
        print(f"Error saving to Google Drive: {e}")

        
# 第二個函式的使用範例
file_name = "活頁簿1.xlsx"
folder_id = "data_upload"
save_to_google_drive(stock_data, file_name, folder_id)

# 顯示 Excel 資料在 GUI 中
show_excel_data_in_gui(file_name)
```
